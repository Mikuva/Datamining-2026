```python
import pandas as pd
import pyczso

# 1. Stáhneme aktuální katalog všech datasetů z API ČSÚ
catalogue = pyczso.czso_get_catalogue()

# Vypíšeme si, kolik datových sad ČSÚ celkem nabízí
print("Celkový počet datasetů v katalogu:", len(catalogue))

# Zobrazíme si prvních 5 řádků, abychom viděli, jak data vypadají
catalogue.head()
```

```python
# 2. Normalizujeme sloupce, aby se v nich dalo pohodlně vyhledávat
katalog_upraveny = pyczso.czso_normalize_catalogue(catalogue)

# 3. Vyfiltrujeme pouze ty datasety, které obsahují klíčové slovo (např. "obyvatel")
nalezena_data = pyczso.czso_filter_catalogue(
    katalog_upraveny,
    text=["obyvatel"],
    latest_only=True
)

print("Počet nalezených datasetů pro náš dotaz:", len(nalezena_data))

# Zobrazíme si jen kódy a názvy prvních 10 nalezených datasetů
nalezena_data[["dataset_code", "dataset_title"]].head(10)
```
