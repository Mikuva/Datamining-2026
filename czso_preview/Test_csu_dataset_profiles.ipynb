# CZSO shortlist – head() + základní charakteristiky

Notebook používá `czso_shortlist_profiles.py`, `pyczso.py` a `pyczso_lkod_bridge.py`.
Pro každý dataset ze shortlistu zobrazí:

- `head()`
- počet řádků a sloupců
- detekované časové sloupce
- roky obsažené v datasetu
- frekvenci a časový rozsah z katalogu
- základní projektová metadata (priorita, skupina, územní úroveň)


```python
from pathlib import Path
import pandas as pd

from czso_shortlist_profiles import display_shortlist_profiles

# Uprav si jen tuto část podle své lokální složky
BASE_DIR = Path.cwd()
CATALOG_XLSX = BASE_DIR / 'czso_katalog_trideny.xlsx'
OUTPUT_DIR = BASE_DIR / 'czso_dataset_profiles'

HEAD_ROWS = 5
MAX_DATASETS = None   # např. 10 pro rychlý test, None = celý shortlist
PRIORITY = None       # např. ['A'] nebo ['A', 'B']; None = vše
PAUSE_SECONDS = 0.0
FORCE_REDOWNLOAD = False

print('Base dir:', BASE_DIR)
print('Catalog :', CATALOG_XLSX)
print('Output  :', OUTPUT_DIR)
```

```python
profiles_df = display_shortlist_profiles(
    catalog_xlsx=CATALOG_XLSX,
    output_dir=OUTPUT_DIR,
    head_rows=HEAD_ROWS,
    max_datasets=MAX_DATASETS,
    priority=PRIORITY,
    pause_seconds=PAUSE_SECONDS,
    force_redownload=FORCE_REDOWNLOAD,
)

print('Hotovo. Výstupní manifest:', OUTPUT_DIR / 'dataset_profiles_manifest.csv')
print('Markdown report       :', OUTPUT_DIR / 'dataset_profiles_report.md')
print('CSV head previews     :', OUTPUT_DIR / 'heads_csv')
```

```python
# Souhrnný pohled na výsledný manifest
profiles_df.head()
```

```python
# Rychlý přehled stavů a pokrytí let
summary = (
    profiles_df
    .groupby(['priority', 'status'], dropna=False)
    .size()
    .reset_index(name='count')
    .sort_values(['priority', 'count'], ascending=[True, False])
)
summary
```
