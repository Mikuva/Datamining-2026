```python
seznam_kraju = [
    "Hlavní město Praha", "Středočeský kraj", "Jihočeský kraj", "Plzeňský kraj", 
    "Karlovarský kraj", "Ústecký kraj", "Liberecký kraj", "Královéhradecký kraj", 
    "Pardubický kraj", "Kraj Vysočina", "Jihomoravský kraj", "Olomoucký kraj", 
    "Zlínský kraj", "Moravskoslezský kraj"]
kraje_geometrie = ox.geocode_to_gdf(seznam_kraju)
Kraje_slovnik = {
 'Central Bohemian Region' : 'Středočeský kraj' ,
 'Hradec Králové Region' : 'Královéhradecký kraj',
 'Karlovy Vary Region' : 'Karlovarský kraj',
 'Liberec Region' : 'Liberecký kraj',
 'Moravian-Silesian Region' : 'Moravskoslezský kraj',
 'Olomouc Region' : 'Olomoucký kraj',
 'Pardubice Region' : 'Pardubický kraj',
 'Plzeň Region' : 'Plzeňský kraj',
 "Prague" : "Hlavní město Praha",
 'South Bohemian Region' : 'Jihočeský kraj',
 'South Moravian Region' : 'Jihomoravský kraj',
 'Vysočina Region' : 'Kraj Vysočina',
 'Zlín Region' : 'Zlínský kraj',
 'Ústí nad Labem Region' : 'Ústecký kraj'}
```

```python
# Získání popisných dat obcí - kód, kraj, název...
Obce_data = requests.get("https://vdp.cuzk.gov.cz/vdp/ruian/obce?sort=NAZEV&ohradaId=&nespravny=&kodVc=&kodOk=&kodOp=&kodPu=&nazevOb=&statusKod=&search=&mediaType=csv")
Obce_data = Obce_data.content.decode('utf-8', errors='replace')
Obce_data = pd.read_csv(io.StringIO(Obce_data), sep=';')
```

```python
# Získání dat o exekucích - kód obce, počet osob v exekuci a počet exekucí
Exekuce_data = requests.get("http://statistiky.ekcr.info/otevrena-data/data/obce.csv")
Exekuce_data = Exekuce_data.content.decode('utf-8', errors='replace')
Exekuce_data = pd.read_csv(io.StringIO(Exekuce_data), sep=',').rename(columns={"pocet_osob" : "Osob v exekuci" })
```

```python
# Získání dat o počtu obyvatel v dané obci, včetně rozložení na pohlaví + průměrný věk
Obyvatele_data = requests.get("https://csu.gov.cz/docs/107508/ebed5ef3-dca1-baf2-c0d3-824b0893086f/1300722503.xlsx?version=1.0")
Obyvatele_data = pd.read_excel(io.BytesIO(Obyvatele_data.content), skiprows=5)
Obyvatele_data = Obyvatele_data.iloc[:,1:]
Obyvatele_data.columns = ["Kód obce", "Obec", "Obyvatel_celkem", "Mužů", "Žen", "Průměrný věk","Průměrný věk muži","Průměrný věk ženy"]
Obyvatele_data = Obyvatele_data.dropna(subset=["Kód obce"])
```

```python
vysledny_df = (
    Obce_data
    .merge(Exekuce_data, left_on='Kód', right_on='kod_obce_zuj', how='inner').drop(columns=["kod_obce_zuj"])
    .merge(Obyvatele_data, left_on='Kód', right_on='Kód obce', how='inner')
    .drop(columns=["Kód obce", "okres","nazev_obce", "Název Kraje (VÚSC)", "Obec", "obvod_orp"])
    .rename(columns={"Kód" : "Kód obce", "Kód Kraje (VÚSC)" : "Kód kraje"}))
```

```python
vysledny_df.head()
```

```python
# Výpočet % exekucí v obcích
obec_exekuce = vysledny_df[['Název Obce', 'Kód obce', "Název Okresu" ,'Osob v exekuci', 'Obyvatel_celkem', "kraj"]].copy()
obec_exekuce['procento_exekuci_obec'] = (obec_exekuce['Osob v exekuci'] / obec_exekuce['Obyvatel_celkem']) * 100
```

```python
# Výpočet % exekucí v jednotlivých okresech
okras_summary = vysledny_df.groupby(['Název Okresu', "Kód Okresu",'kraj'])[['Osob v exekuci', 'Obyvatel_celkem']].sum().reset_index()
okras_summary['procento_exekuci_okres'] = (okras_summary['Osob v exekuci'] / okras_summary['Obyvatel_celkem']) * 100
okres_exekuce = okras_summary[['Název Okresu', "Kód Okresu", 'procento_exekuci_okres','kraj',"Osob v exekuci","Obyvatel_celkem"]]
```

```python
# Výpočet % exekucí v jednotlivých krajích
kraje_summary = vysledny_df.groupby(["kraj",'Kód kraje'])[['Osob v exekuci', 'Obyvatel_celkem']].sum().reset_index()
kraje_summary['procento_exekuci_kraj'] = (kraje_summary['Osob v exekuci'] / kraje_summary['Obyvatel_celkem']) * 100
kraj_exekuce = kraje_summary[["kraj",'Kód kraje', 'procento_exekuci_kraj',"Osob v exekuci","Obyvatel_celkem"]]
```

```python
obec_exekuce.sort_values(by="procento_exekuci_obec", ascending=False)
```
